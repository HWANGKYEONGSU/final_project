```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data/Byeonghap_final4.csv", index_col = 0)
```

```python
df = pd.DataFrame(data)
df
```

```python
df=df.reset_index()
```

```python
df
```

```python
# 결과 컬럼 추가 : 성사 - 1 , 실패 - 0

df["result"] = 3   # 아무값이나 해서 컬럼 생성

for i in range(len(df["percentage"])):
    if df["percentage"][i] >= 100:
        df["result"][i] = 1
    else: 
        df["result"][i] = 0
```

```python
df
```

```python
df.to_csv("data/df_reset.csv", encoding = "utf-8")
```

```python
X = df.category.values
y = df.result.values
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y, alpha=0.5)
plt.title("category ~ result")
plt.xlabel("category")
plt.ylabel("result")
plt.show()
```
