# 목표 금액 상관분석 

```python
data_set = df_normed[["target_funding_amount","result"]]
data_set
```

```python
plt.figure(figsize=(20,5))

plt.scatter(data_set["target_funding_amount"],data_set["result"], color="red")

plt.show()
```

```python
# 피어슨 상관관계
rho, p_val = stats.pearsonr(data_set["target_funding_amount"],data_set["result"])

print("correlation coefficient : {}, p-value:{}".format(rho, p_val))
```

```python
# 스피어만 상관관계
rho, p_val = stats.spearmanr(data_set["target_funding_amount"],data_set["result"])

print("correlation coefficient : {}, p-value:{}".format(rho, p_val))
```

```python
# 켄달타우 상관관계
rho, p_val = stats.kendalltau(data_set["target_funding_amount"],data_set["result"])

print("correlation coefficient : {}, p-value:{}".format(rho, p_val))
```

# 목표 금액 회귀 분석

```python
# 독립변수 : X 

X = data_set[['target_funding_amount']]
print(type(X)) 


# 종속변수 : y 

y = data_set['result']
print(type(y))


```

```python
# 데이터셋 구분 - 훈련용(train data) / 검증용(test data)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=10)
print('train data 개수: ', len(X_train))
print('test data 개수: ', len(X_test))
```

```python
# 학습 데이터
print(X_train.shape)
print(y_train.shape)

#테스트 데이터
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
# train data를 가지고 모형 학습

# X_train : 훈련데이터. y_train : 정답

lr.fit(X_train, y_train)
```

```python
# 결정계수(R 제곱)

r_square = lr.score(X_test, y_test)
print(r_square)
```

```python
# 기울기

print("기울기 a:", lr.coef_)
```

```python
# y 절편

print("y 절편 b:", lr.intercept_)
```

```python
# 테스트하기. 전체 데이터 사용.
# pred : 예측한 값

pred = lr.predict(X)
```

```python
# 원래 값 y와 예측된 값 pred 데이터를 그래프로 작성하기


plt.figure(figsize=(10,5))
ax1 = sns.kdeplot(y, label="y")                   # 실제 result 데이터의 그래프(정답). kdeplot : 빈도수, 밀도에 해당
ax2 = sns.kdeplot(pred, label='pred', ax=ax1)   # 예측한 값 
plt.legend()
plt.show()
```
