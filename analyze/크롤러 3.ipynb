```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0'}
failed = []
project_url = []
intro = []
category = []
founder = []

# "펀딩 무산. 목표 금액을 달성하지 못했습니다. 결제가 진행되지 않습니다."하는 검색의 30페이지(끝)까지 수집 
for page in range(0, 290, 10):
    URL = 'https://www.google.com/search?q=%ED%85%80%EB%B8%94%EB%B2%85+%ED%8E%80%EB%94%A9+%EB%AC%B4%EC%82%B0.+%EB%AA%A9%ED%91%9C+%EA%B8%88%EC%95%A1%EC%9D%84+%EB%8B%AC%EC%84%B1%ED%95%98%EC%A7%80+%EB%AA%BB%ED%96%88%EC%8A%B5%EB%8B%88%EB%8B%A4.+%EA%B2%B0%EC%A0%9C%EA%B0%80+%EC%A7%84%ED%96%89%EB%90%98%EC%A7%80+%EC%95%8A%EC%8A%B5%EB%8B%88%EB%8B%A4.&ei=RJLFYaC-MqeWr7wPvbSWmAU&start='+str(page)+'&sa=N&filter=0&ved=2ahUKEwig8IqZj_z0AhUny4sBHT2aBVM4ChDy0wN6BAgBED0&biw=1536&bih=722&dpr=1.25'
    res = requests.get(URL, headers=headers)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    for i in range(10): # 한 페이지에 프로젝트 10개씩 있음 
        failed.append(soup.find_all("h3")[i].text) 
        project_url.append(soup.find_all("div", "kCrYT")[i*2].a["href"][7:])
        
print(len(failed))
print(len(project_url))
```
