```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
name_list = []
price_list =[]
money_list = []
founder_list =[]

driver = webdriver.Chrome("C:/Users/admin/chromedriver")

for i in df["url"][:2]:    
    url = i
    driver.get(url)

    name = driver.find_elements_by_css_selector("div.ProjectIntroduction__ProjectIntroductionBackground-sc-1o2ojgb-1.inZhnj > div > div > div.ProjectIntroduction__ProjectOutline-sc-1o2ojgb-3.jYmvPn > div > h1")
    price_b = driver.find_elements_by_css_selector("div.ProjectIntroduction__ProjectIntroductionBackground-sc-1o2ojgb-1.inZhnj > div > div > aside > div.ProjectIntroduction__FundingStatus-sc-1o2ojgb-14.cAPmYx > div:nth-child(1) > div.ProjectIntroduction__StatusValue-sc-1o2ojgb-17.jqaUQg")
    price = []
    money = driver.find_elements_by_css_selector("div.ProjectIntroduction__ProjectIntroductionBackground-sc-1o2ojgb-1.inZhnj > div > div > aside > div.FundingInformation-cjd67l-0.lfNPLI > div > span")
    founder = []
    founder_b = driver.find_elements_by_css_selector("div.ProjectIntroduction__ProjectIntroductionBackground-sc-1o2ojgb-1.inZhnj > div > div > aside > div.ProjectIntroduction__FundingStatus-sc-1o2ojgb-14.cAPmYx > div:nth-child(3) > div.ProjectIntroduction__StatusValue-sc-1o2ojgb-17.jqaUQg")

    for i in range(len(price_b)):
        price1 = price_b[i].text
        price_list.append(price1)

    for i in name:
        name1 = i.text
        name_list.append(name1)

        
    for i in money:
        money1 = i.text
        money_list.append(money1)
        
        
    for i in range(len(founder_b)):
        founder1 = founder_b[i].text
        founder_list.append(founder1)
    
    
driver.close()
```

```python
import re

money_list_cut = []

for i in money_list:


    try:
        found = re.search('목표 금액인 (.+?)원이', i).group(1)
    except AttributeError:
        pass
    money_list_cut.append(found)
    


price_list_per = []

for i in price_list:


    try:
        found = re.search('원(.+?)%', i).group(1)
    except AttributeError:
        pass
    price_list_per.append(found)

price_list_cost = []

for i in price_list:


    try:
        found = re.search('(.+?)원', i).group(1)
    except AttributeError:
        pass
    price_list_cost.append(found)
    
    
```

```python
founder_list
```

```python
data = {"상품명":name_list,
        "현재 펀딩 금액": price_list_cost,
         "펀딩 퍼센트": price_list_per,
        "목표금액": money_list_cut,
        "후원자 수": founder_list
       }

df2 = pd.DataFrame(data)
df2
```

### 4. 크롤링한 df2와 기존 df(Not_Jinhaeng_join)을 병합

```python
df["현재 펀딩 금액"] = df2["현재 펀딩 금액"]
df["펀딩 퍼센트"] = df2["펀딩 퍼센트"]
df["목표 금액"] = df2["목표 금액"]
df["후원자 수"] = df2["후원자 수"]
```
