```python
new_wish = df1["현재 펀딩 금액"] / df1["펀딩 퍼센트"]

for i in range(len(df1["목표금액"].values)):         # i는 순번
    if df1["목표금액"].values[i] == "배송 완료":
        df1["목표금액"].values[i] = new_wish.values[i].astype('int')
    else: 
        pass

df1["목표금액"] = df1["목표금액"].values
```
